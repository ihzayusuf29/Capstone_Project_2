# `Latar Belakang`
Lazada merupakan sebuah e-commerce besar di indonesia yang bergerak di bidang online retail, menurut situs [goodstats.id](https://goodstats.id/), lazada berada pada peringkat ketiga sebagai ecommerce terpopuler di indonesia, dengan jumlah kunjungan hingga 26,64 juta kali pada kuartal ii tahun 2022 dan 74,5 juta kali pada kuartal ii tahun 2023.

sebagai sebuah e-commerce besar, lazada selalu mencoba untuk terus berinovasi dan mengevaluasi kinerjanya, lazada ingin meningkatkan efektivitas dan efisiensi proses bisnisnya, termasuk juga tingkat efektivitas pengiriman produk hingga ke tangan customer.
